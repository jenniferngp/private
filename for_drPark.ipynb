```r
# open file
data = fread("/frazer01/projects/CEGS/analysis/ukbb.gwas.kidney/pipeline/gwas/sumstats/standard.f.130708.0.0.AFR.txt", data.table = F)
head(data,2)
```

```r
library(dplyr)

# arrange by pvalue from small to large
data = data %>% arrange(pval)

# print first 100 rows
head(data,10)

# filter only significant results by p-value < 0.05
# method 1: using dplyr
filtered_data = data %>% filter(pval <= 0.05) 
# method 2: using base R
filtered_data = data[data$pval <= 0.05,]

# to count how many rows of filtered data
nrow(filtered_data)

# plot full data
library(ggplot2)
ggplot(filtered_data, aes(x = pos, y = -log10(pval))) + geom_point()

```

```r
# filter data by chr15 
filt_data_15 = data %>% filter(chrom == 15 & pval <= 0.05)

summary(filt_data_15$pval)

ggplot(filt_data_15, aes(x = pos, y = -log10(pval))) + 
    geom_point(size = 3) + 
    theme(axis.text = element_text(size = 15),
          axis.title = element_text(size = 15)) 
```
